```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned/superstore_enriched.csv")
```

```python
customers = df[
    [
        "Customer ID",
        "Customer Name",
        "Segment"
    ]
].drop_duplicates().reset_index(drop=True)
```

```python
customers.columns = [
    "customer_id",
    "customer_name",
    "segment"
]
```

```python
customers.head()
```

```python
customers.shape
```

```python
customers.duplicated().sum()
```

```python
customers.to_csv(
    "../data/normalized/customers.csv",
    index=False
)

print("customers.csv created successfully!")
```

```python
products = df[
    [
        "Product ID",
        "Product Name",
        "Category",
        "Sub-Category"
    ]
].drop_duplicates().reset_index(drop=True)
```

```python
products.columns = [
    "product_id",
    "product_name",
    "category",
    "sub_category"
]
```

```python
products.head()
```

```python
products.shape
```

```python
products.duplicated().sum()
```

```python
products.to_csv(
    "../data/normalized/products.csv",
    index=False
)

print("products.csv created successfully!")
```

```python
locations = df[
    [
        "Country",
        "Region",
        "State",
        "City",
        "Postal Code"
    ]
].drop_duplicates().reset_index(drop=True)
```

```python
locations.columns = [
    "country",
    "region",
    "state",
    "city",
    "postal_code"
]
```

```python
locations.insert(
    0,
    "location_id",
    range(1, len(locations) + 1)
)
```

```python
locations.head()
```

```python
locations.shape
```

```python
locations.to_csv(
    "../data/normalized/locations.csv",
    index=False
)

print("locations.csv created successfully!")
```

```python
orders = df.copy()
```

```python
orders = orders[
    [
        "Order ID",
        "Order Date",
        "Ship Date",
        "Ship Mode",
        "Customer ID",
        "Product ID",
        "Country",
        "Region",
        "State",
        "City",
        "Postal Code",
        "Sales",
        "Quantity",
        "Discount",
        "Profit"
    ]
]
```

```python
orders.columns = [
    "order_id",
    "order_date",
    "ship_date",
    "ship_mode",
    "customer_id",
    "product_id",
    "country",
    "region",
    "state",
    "city",
    "postal_code",
    "sales",
    "quantity",
    "discount",
    "profit"
]
```

```python
orders.head()
```

```python
orders.shape
```

```python
orders = orders.merge(
    locations,
    on=["country", "region", "state", "city", "postal_code"],
    how="left"
)
```

```python
orders["location_id"].isnull().sum()
```

```python
orders = orders[
    [
        "order_id",
        "order_date",
        "ship_date",
        "ship_mode",
        "customer_id",
        "product_id",
        "location_id",
        "sales",
        "quantity",
        "discount",
        "profit"
    ]
]
```

```python
orders.head()
```

```python
orders.shape
```

```python
orders.to_csv(
    "../data/normalized/orders.csv",
    index=False
)

print("orders.csv created successfully!")
```
